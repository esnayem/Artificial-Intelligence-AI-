### Imports Library

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Define a,b & x

```python
a, b = 5, -11
x = np.array(range(-10,11))
```

### Equation Y = ax + b

```python
y = a * x + b
plt.plot(x,y, linestyle = '--', marker = 'o', color = 'yellow', label = 'line with marker')
for i in range(len(x)):
    plt.annotate(f"({x[i]}, {y[i]})",(x[i],y[i]))
plt.title("Y = aX + b")
plt.legend()
```

### Equation y = asin(x) + b

```python
y = a * np.sin(x) + b
plt.plot(x,y, linestyle = '--', marker = 'o', color = 'yellow', label = 'line with marker')
for i in range(len(x)):
    plt.annotate(f"({np.round(x[i],3)}, {np.round(y[i],3)})",(x[i],y[i]))
plt.title("Y = asin(x) + b")
plt.legend()
```

### Equation y = ax^2 + b

```python
y = a * x * x + b
plt.plot(x,y, linestyle = '--', marker = 'o', color = 'yellow', label = 'line with marker')
for i in range(len(x)):
    plt.annotate(f"({np.round(x[i],3)}, {np.round(y[i],3)})",(x[i],y[i]))
plt.title("Y = ax^2 + b")
plt.legend()
```

### Equation y = ae^x + b

```python
y = a * np.exp(x) + b
plt.plot(x,y, linestyle = '--', marker = 'o', color = 'yellow', label = 'line with marker')
for i in range(len(x)):
    plt.annotate(f"({np.round(x[i],3)}, {np.round(y[i],3)})",(x[i],y[i]))
plt.title("Y = ae^x + b")
plt.legend()
```
